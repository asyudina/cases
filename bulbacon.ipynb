```python
import random
import itertools
```

```python
#set probabilities of directions
```

```python
p_up = 0.3*0.4
p_down = 0.2*0.4
p_lft = 0.4*0.5
p_rgt = 0.2*0.5

p_l_up = 0.3* 0.4
p_l_down = 0.2 * 0.4
p_r_up=0.3*0.2
p_r_down=0.2*0.2
p_stay=0.4*0.5
```

```python
#set function of detecting position after moving. parameters - initial position, direction 
def moves(init,mv):
    if mv == 'up':
        init[0]+=1
    elif mv == 'l_up':
        init[0]+=1
        init[1]-=1
    elif mv == 'lft':
        init[1]-=1
    elif mv == 'l_down':
        init[0]-=1
        init[1]-=1
    elif mv == 'down':
        init[0]-=1
    elif mv == 'r_down':
        init[0]-=1
        init[1]+=1
    elif mv == 'rgt':
        init[1]+=1
    elif mv == 'r_up':
        init[0]+=1
        init[1]+=1
    elif mv == 'stay':
        init = init
    return(init)
```

```python
#set where king can move and probabilities of moving in this direction 
movements = ['up', 'l_up','lft','l_down','down','r_down','rgt','r_up','stay']
m_weights = [p_up,p_l_up,p_lft,p_l_down,p_down,p_r_down,p_rgt,p_r_up,p_stay]

#probability to be in initial position after 5 rounds
total_prob = 0


#for each combination of directions check position after 5 rounds, 
#if it is initial calculate probability of this set of directions and add to final probability
for subset in itertools.product(movements, repeat=5):
    init = [0,0]
    prob = 1
    for movement in subset:
        moves(init,movement)
        pos = movements.index(movement)
        curr_prob = m_weights[pos]
        prob = prob*curr_prob
    final_pos = init
    if final_pos ==[0,0]:
        print(subset,final_pos,prob)
        total_prob+=prob
    else:
        continue

    
```

```python
#answer 
print(total_prob) 
```
